```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_node('Mathilda')
G.add_node('Sacha')

G.add_edge('Killian', 'Celian') # Network X will check if those two nodes exist, if not, it will draw a connection between them and create the nodes by default.

# G.add_edges_from([('Mathilda', 'Sacha')])

nx.draw(G)
plt.show()
```

```python
G2 = nx.Graph()

G2.add_node('Victor')

names = [('Mathilda', 'Sacha'),('Killian', 'Celian'), ('Killian', 'Mathilda'), ('Killian', 'Sacha'), ('Mathilda', 'Victor')]

G2.add_edges_from(names)

G2.remove_edge('Mathilda', 'Victor')

# G2.clear()

print(nx.number_of_edges(G2))

nx.draw(G2, with_labels=True, font_family='Times New Roman', font_size=8, font_color='blue', bbox = dict(facecolor='red', alpha=.5))
plt.show()
```
